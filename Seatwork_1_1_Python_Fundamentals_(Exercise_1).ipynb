```python
def calculate_grade(exam, hoa, quiz, assignment, weight_exam, weight_hoa, weight_quiz, weight_assignment):
    """Calculates a grade component (CS or Grade) based on given scores and weights."""
    class_standing = (weight_hoa * hoa) + (weight_quiz * quiz) + (weight_assignment * assignment)
    grade = (weight_exam * exam) + (weight_exam * class_standing)  # Assuming exam weight is also 50%
    return grade


def calculate_weighted_average(grade1, grade2, weight1, weight2):
    """Calculates a weighted average of two grades."""
    return (weight1 * grade1) + (weight2 * grade2)


def convert_to_tip_grade(final_grade):
    """Converts the final grade to the T.I.P. numerical grade equivalent."""
    if 94 <= final_grade <= 100:
        return 1.00
    elif 88.5 <= final_grade <= 93.99:
        return 1.25
    elif 83 <= final_grade <= 88.49:
        return 1.50
    elif 77.5 <= final_grade <= 82.99:
        return 1.75
    elif 72 <= final_grade <= 77.49:
        return 2.00
    elif 65.5 <= final_grade <= 71.99:
        return 2.25
    elif 61 <= final_grade <= 65.49:
        return 2.50
    elif 55.5 <= final_grade <= 60.99:
        return 2.75
    elif 50 <= final_grade <= 55.49:
        return 3.00
    elif 0 <= final_grade <= 49.99:
        return 5.00  # Assuming 5.00 for Failed as per the table (0-49.99)
    else:
        return "Invalid Grade"  # Handle cases outside the defined range


def get_descriptive_rating(final_grade):
    """Gets the descriptive rating based on the final grade."""
    if 90 <= final_grade <= 100:
        return "Excellent"
    elif 85 <= final_grade <= 89.99:
        return "Superior"
    elif 80 <= final_grade <= 84.99:
        return "Meritorious"
    elif 75 <= final_grade <= 79.99:
        return "Very Good"
    elif 70 <= final_grade <= 74.99:
        return "Good"
    elif 65 <= final_grade <= 69.99:
        return "Very Satisfactory"
    elif 60 <= final_grade <= 64.99:
        return "Satisfactory"
    elif 55 <= final_grade <= 59.99:
        return "Fair"
    elif 50 <= final_grade <= 54.99:
        return "Passing"
    elif 40 <= final_grade <= 49.99:  # Added a range for Incomplete based on your table
        return "Incomplete"
    elif 0 <= final_grade <= 39.99:  # Adjusted range for Failed.
        return "Failed"
    else:
        return "Invalid Grade"
        # --- Main program ---

num_students = int(input("Enter the number of students: "))  # Get the number of students from the user

for student in range(num_students):
    print(f"\n--- Student {student+1} ---")
    name = input("Enter student name: ")

    # Get input for grades
    prelim_exam = float(input("Enter prelim exam score: "))
    prelim_hoa = float(input("Enter average prelim HOA score: "))
    prelim_quiz = float(input("Enter average prelim quiz score: "))
    prelim_assignment = float(input("Enter average prelim assignment score: "))

    midterm_exam = float(input("Enter midterm exam score: "))
    midterm_hoa = float(input("Enter average midterm HOA score: "))
    midterm_quiz = float(input("Enter average midterm quiz score: "))
    midterm_assignment = float(input("Enter average midterm assignment score: "))

    final_exam = float(input("Enter final exam score: "))
    final_hoa = float(input("Enter average final HOA score: "))
    final_quiz = float(input("Enter average final quiz score: "))
    final_assignment = float(input("Enter average final assignment score: "))

    # Calculate Prelim Grade
    prelim_grade = calculate_grade(prelim_exam, prelim_hoa, prelim_quiz, prelim_assignment, 0.5, 0.5, 0.3, 0.2)

    # Calculate Midterm Grade
    midterm_cs = calculate_grade(midterm_exam, midterm_hoa, midterm_quiz, midterm_assignment, 0.5, 0.5, 0.3, 0.2)
    midterm_grade = calculate_weighted_average(prelim_grade, midterm_cs, 1/3, 2/3)

    # Calculate Final Grade
    final_cs = calculate_grade(final_exam, final_hoa, final_quiz, final_assignment, 0.5, 0.5, 0.3, 0.2)
    final_grade = calculate_weighted_average(midterm_grade, final_cs, 1/3, 2/3)

    # Output the results
    print("\n--- Final Grade Calculation ---")
    print("Prelim Grade:", prelim_grade)
    print("Midterm Grade:", midterm_grade)
    print("Final Grade:", final_grade)

    tip_grade = convert_to_tip_grade(final_grade)
    print("T.I.P. Grade:", tip_grade)

    descriptive_rating = get_descriptive_rating(final_grade)
    print("Descriptive Rating:", descriptive_rating)
```

```python
def calculate_prelim_grade(prelim_exam, prelim_cs):
  """Calculates the prelim grade."""
  return 0.5 * prelim_exam + 0.5 * prelim_cs

def calculate_midterm_grade(prelim_grade, midterm_exam, midterm_cs):
  """Calculates the midterm grade."""
  return (1/3) * prelim_grade + (2/3) * (0.5 * midterm_exam + 0.5 * midterm_cs)

def calculate_final_grade(midterm_grade, final_exam, final_cs):
  """Calculates the final grade."""
  return (1/3) * midterm_grade + (2/3) * (0.5 * final_exam + 0.5 * final_cs)

def calculate_class_standing(hoa_avg, quiz_avg, assignment_avg):
  """Calculates the class standing score."""
  return 0.5 * hoa_avg + 0.3 * quiz_avg + 0.2 * assignment_avg


# --- Main program ---

# Input Prelim Scores
prelim_hoa = float(input("Enter Prelim HOA average: "))
prelim_quiz = float(input("Enter Prelim Quiz average: "))
prelim_assignment = float(input("Enter Prelim Assignment average: "))
prelim_exam = float(input("Enter Prelim Exam score: "))

# Calculate Prelim Class Standing and Grade
prelim_cs = calculate_class_standing(prelim_hoa, prelim_quiz, prelim_assignment)
prelim_grade = calculate_prelim_grade(prelim_exam, prelim_cs)

# Input Midterm Scores
midterm_hoa = float(input("Enter Midterm HOA average: "))
midterm_quiz = float(input("Enter Midterm Quiz average: "))
midterm_assignment = float(input("Enter Midterm Assignment average: "))
midterm_exam = float(input("Enter Midterm Exam score: "))

# Calculate Midterm Class Standing and Grade
midterm_cs = calculate_class_standing(midterm_hoa, midterm_quiz, midterm_assignment)
midterm_grade = calculate_midterm_grade(prelim_grade, midterm_exam, midterm_cs)

# Input Final Scores
final_hoa = float(input("Enter Final HOA average: "))
final_quiz = float(input("Enter Final Quiz average: "))
final_assignment = float(input("Enter Final Assignment average: "))
final_exam = float(input("Enter Final Exam score: "))

# Calculate Final Class Standing and Grade
final_cs = calculate_class_standing(final_hoa, final_quiz, final_assignment)
final_grade = calculate_final_grade(midterm_grade, final_exam, final_cs)

# Output the Final Grade
print("Final Grade:", final_grade)
```
